```python
#graph 1
fig, ax1 = plt.subplots()
fig.tight_layout()
fig.set_size_inches(13, 6, forward=True)

lns1 = ax1.plot(get_num_stories_df(fdata), color='r', label='Story Count from WSJ')
ax1.set_title("Bitcoin/Cryptocurrency Stories", size = 20) 
ax1.set_ylabel('Total Number of Stories')
ax1.set_xlabel("")


ax2 = ax1.twinx() 
lns2 = ax2.plot(yf.Ticker("BTC-USD").history(period="max")["Close"], color='b', label="Bitcoin Price (USD)")
ax2.set_ylabel('Price in USD')

lns = lns1+lns2
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, loc=0)

plt.show()
```
